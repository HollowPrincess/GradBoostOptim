```python
#  Pick a Dataset you might be interested in.
#  Say, all airline-safety files...
import zipfile

file = "/kaggle/input/liberty-mutual-group-property-inspection-prediction/train.csv.zip"

# Will unzip the files so that you can see them..
with zipfile.ZipFile(file,"r") as z:
    z.extractall(".")
```

```python
'''


Based on Abhishek Catapillar benchmark
https://www.kaggle.com/abhishek/caterpillar-tube-pricing/beating-the-benchmark-v1-0

@Soutik

Have fun;)
'''

"""
Взята предобработка данных для модели
"""

from sklearn import preprocessing
import xgboost as xgb
from sklearn.feature_extraction import DictVectorizer



train  = pd.read_csv("./train.csv", index_col=0)
print("Dataset shape: ", train.shape)
labels = train.Hazard
train.drop('Hazard', axis=1, inplace=True)
train = train.T.to_dict().values()
vec = DictVectorizer()
train = vec.fit_transform(train)
```

```python
print("Processed data shape:", train.shape)
```

```python
train = pd.DataFrame(train.toarray())
train["labels"] = labels.values
```

```python
train.to_csv("prepared_prop_inspection.csv", index=False)
```
